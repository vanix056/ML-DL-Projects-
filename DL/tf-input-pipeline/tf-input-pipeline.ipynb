```python
class_names = ['positive', 'negative']
```

```python
def get_label(file_path):
    import os
    parts = tf.strings.split(file_path, os.path.sep)
    return parts[-2]
```

```python
get_label('NLP Stuff\\reviews\\negative\\neg_1.txt')
```

```python
def process_txt(file_path):
    label = get_label(file_path)
    review_text = tf.io.read_file(file_path) # load the raw data from the file as a string
    return review_text, label
```

```python
txt, label = process_txt("NLP Stuff\\reviews\\positive\\pos_2.txt")
print(txt.numpy())
print(label.numpy())
```

```python
Comp_ds = txt_ds.map(process_txt)
for txt, label in Comp_ds:
    print("****",txt.numpy())
    print("****",label.numpy())
```

```python
len(Comp_ds)
```

```python
final_ds = Comp_ds.filter(lambda txt, label: tf.strings.length(txt) > 0)
```

```python
for txt, label in final_ds.take(5):
    print("****Text: ",txt.numpy())
    print("****Label: ",label.numpy())
```

```python
sing_txt_ds = tf.data.Dataset.list_files("NLP Stuff/reviews/*/*")
sing_txt_ds = sing_txt_ds.map(process_txt).filter(lambda txt, label: tf.strings.length(txt) > 0).shuffle(4)
for txt, label in sing_txt_ds:
    print("****Text: ",txt.numpy())
    print("****Label: ",label.numpy())
```

```python
import numpy as np
```

```python
y_predicted = np.array([1,1,0,0,1])
y_true = np.array([0.30,0.7,1,0,0.5])
```

```python
def mse(y_predicted,y_true):
    totalerror = 0
    for yp,yt in zip(y_predicted,y_true):
        totalerror += (yt-yp)**2
    print(f'Total Error is {totalerror}')
    mse=totalerror/len(y_true)
    print(f'Total MSE is {mse}')
    return mse
```

```python
mse(y_predicted, y_true)
```

```python
# Now by using numpy
def mse_num(y_predicted,y_true):
    return np.mean(np.square(y_true-y_predicted))
```

```python
mse_num(y_predicted, y_true)
```

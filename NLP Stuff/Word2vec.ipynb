# dataset used :

```python
import pandas as pd
import gensim
```

```python
df=pd.read_json("reviews_Sports_and_Outdoors_5_extracted.json", lines=True)
df
```

```python
review_text = df.reviewText.apply(gensim.utils.simple_preprocess)
review_text
```

```python
review_text.iloc[1]
```

```python
df.reviewText.iloc[1]
```

```python
model = gensim.models.Word2Vec(
    window=5,
    min_count=2,
    workers=4,

)
```

```python
model.build_vocab(review_text, progress_per=1000)
```

```python
model.train(review_text, total_examples=model.corpus_count, epochs=model.epochs)
```

# Testing the Similarity between Words

```python
model.wv.most_similar("bad")
```

```python
model.wv.similarity("great", "good")
```

```python
model.wv.similarity("slow", "steady")
```

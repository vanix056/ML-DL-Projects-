```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv('titanic.csv')
df.head()
```

```python
le=LabelEncoder()
df['sex_n']=le.fit_transform(df.Sex)
df
```

```python
df.drop('Sex', axis='columns', inplace=True)
df# Optional: inplace=True to modify df directly
```

```python
import math
avg=math.floor(df.Age.mean())
avg
```

```python
df['Age']=df.Age.fillna(avg)
df
```

```python
df.head(10)
```

```python
input = df[['Pclass', 'sex_n', 'Age', 'Fare']]
input
```

```python
target=df['Survived']
target
```

```python
from sklearn.model_selection import train_test_split
from sklearn import tree
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(input,target,test_size=0.2)
model=tree.DecisionTreeClassifier(criterion='log_loss')
```

```python
model.fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

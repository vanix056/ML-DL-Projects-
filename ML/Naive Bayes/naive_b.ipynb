```python
import pandas as pd 
from sklearn.datasets import load_wine
```

```python
wine=load_wine()
```

```python
dir(wine)
```

```python
df=pd.DataFrame(wine.data,columns=wine.feature_names)
df
```

```python
df['target']=wine.target
df
```

```python
df['target_names']=df.target.apply(lambda x : wine.target_names[x])
df
```

```python
x=df.drop(['target_names','target'],axis='columns')
x
```

```python
y=df.target
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
```

```python
model1=GaussianNB()
model2=MultinomialNB()
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
model1.fit(xtrain,ytrain)
model2.fit(xtrain,ytrain)
```

```python
model1.score(xtest,ytest)
```

```python
model2.score(xtest,ytest)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(model1,xtest,ytest,cv=6)
```

```python
cross_val_score(model2,xtest,ytest,cv=6)
```

```python
model1.predict_proba(xtest)
```

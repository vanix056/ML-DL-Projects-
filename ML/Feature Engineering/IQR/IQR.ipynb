```python
import pandas as pd
df=pd.read_csv('weight-height.csv')
df
```

```python
import matplotlib.pyplot as plt
plt.hist(df.Height,bins=20,rwidth=0.8)
plt.xlabel('height')
plt.ylabel('count')
```

```python
import matplotlib.pyplot as plt
plt.hist(df.Weight,bins=20,rwidth=0.8)
plt.xlabel('weight')
plt.ylabel('count')
```

```python
IQR1=df.Height.quantile(0.75)-df.Height.quantile(0.25)
min1=df.Height.quantile(0.25)-1.5*IQR1
max1=df.Height.quantile(0.75)+1.5*IQR1
min1,max1
```

```python
IQR2=df.Weight.quantile(0.75)-df.Weight.quantile(0.25)
min2=df.Weight.quantile(0.25)-1.5*IQR2
max2=df.Weight.quantile(0.75)+1.5*IQR2
min2,max2
```

```python
new_df=df[((df.Height>min1)& (df.Height<max1))|((df.Weight>min2)&(df.Weight<max2))]
new_df
```

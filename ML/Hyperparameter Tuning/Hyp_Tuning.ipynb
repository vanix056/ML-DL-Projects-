```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_digits
from sklearn.model_selection import GridSearchCV
```

```python
digit=load_digits()
```

```python
dir(digit)
```

```python
import pandas as pd
df=pd.DataFrame(digit.data,columns=digit.feature_names)
df
```

```python
df['target']=digit.target
df
```

```python
model_params = {
    'svm': {
        'model': svm.SVC(gamma='auto'),
        'params': {
            'C': [1, 10, 20],
            'kernel': ['rbf', 'linear']
        }
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params': {
            'n_estimators': [1, 5, 10]
        }
    },
    'logistic_regression': {
        'model': LogisticRegression(solver='liblinear', multi_class='auto'),
        'params': {
            'C': [1, 5, 10]
        }
    },
    'gaussian_nb': {
        'model': GaussianNB(),
        'params': {}  # No hyperparameters to tune commonly
    },
    'multinomial_nb': {
        'model': MultinomialNB(),
        'params': {
            'alpha': [0.5, 1.0, 1.5]
        }
    },
    'decision_tree': {
        'model': DecisionTreeClassifier(),
        'params': {
            'max_depth': [None, 5, 10],
            'criterion': ['gini', 'entropy']
        }
    }
}
```

```python
score=[]
cv_results_list = []
for models,mp in model_params.items():
    clf=GridSearchCV(mp['model'],mp['params'],cv=5,return_train_score=False)
    clf.fit(digit.data,digit.target)
    score.append({
        'model':models,
        'best_params':clf.best_params_,
        'best_score':clf.best_score_
    })
     #Add detailed cv_results_
    results_df = pd.DataFrame(clf.cv_results_)
    results_df['model'] = models  # Add model name column
    cv_results_list.append(results_df)


df=pd.DataFrame(score,columns=['model','best_score','best_params'])
df
```

```python
# Combine all CV results into a single DataFrame
cv_details = pd.concat(cv_results_list, ignore_index=True)
cv_details
```

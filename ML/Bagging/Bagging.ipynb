```python
import pandas as pd
df=pd.read_csv('heart.csv')
df
```

```python
df.describe()
```

```python
zscore1=(df.RestingBP-df.RestingBP.mean())/df.RestingBP.std()
zscore2=(df.Cholesterol-df.Cholesterol.mean())/df.Cholesterol.std()
zscore3=(df.MaxHR-df.MaxHR.mean())/df.MaxHR.std()
zscore4=(df.Oldpeak-df.Oldpeak.mean())/df.Oldpeak.std()
```

```python
df[['zscore1','zscore2','zscore3','zscore4']] = pd.concat([zscore1, zscore2, zscore3, zscore4], axis=1).values
df
```

```python
df1 = df[
    (df.zscore1 > -3) & (df.zscore1 < 3) &
    (df.zscore2 > -3) & (df.zscore2 < 3) &
    (df.zscore3 > -3) & (df.zscore3 < 3) &
    (df.zscore4 > -3) & (df.zscore4 < 3)
]
df1
```

```python
x=df.drop((['HeartDisease','zscore1','zscore2','zscore3','zscore4']),axis=1)
x
```

```python
y=df.HeartDisease
y
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dfle=df1
dfle.Sex=le.fit_transform(dfle.Sex)
dfle.ChestPainType=le.fit_transform(dfle.ChestPainType)
dfle.RestingECG=le.fit_transform(dfle.RestingECG)
dfle.ExerciseAngina=le.fit_transform(dfle.ExerciseAngina)
dfle.ST_Slope=le.fit_transform(dfle.ST_Slope)
dfle
```

```python
x=df1.drop((['HeartDisease','zscore1','zscore2','zscore3','zscore4']),axis=1)
x
```

```python
y=df1.HeartDisease
y
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(x)
X_scaled
```

```python
df1
```

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

model_params = {
    'svm': {
        'model': svm.SVC(gamma='auto'),
        'params': {
            'C': [1, 10, 20],
            'kernel': ['rbf', 'linear']
        }
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params': {
            'n_estimators': [1, 5, 10]
        }
    },
    'logistic_regression': {
        'model': LogisticRegression(solver='liblinear', multi_class='auto'),
        'params': {
            'C': [1, 5, 10]
        }
    },
    'decision_tree': {
        'model': DecisionTreeClassifier(),
        'params': {
            'max_depth': [None, 5, 10],
            'criterion': ['gini', 'entropy']
        }
    }
}
```

```python
score=[]
cv_results_list = []
for models,mp in model_params.items():
    clf=GridSearchCV(mp['model'],mp['params'],cv=5,return_train_score=False)
    clf.fit(X_scaled,y)
    score.append({
        'model':models,
        'best_params':clf.best_params_,
        'best_score':clf.best_score_
    })
     #Add detailed cv_results_
    results_df = pd.DataFrame(clf.cv_results_)
    results_df['model'] = models  # Add model name column
    cv_results_list.append(results_df)


df=pd.DataFrame(score,columns=['model','best_score','best_params'])
df
```

```python
cv_details = pd.concat(cv_results_list, ignore_index=True)
cv_details
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
from sklearn import linear_model
```

```python
def gradient_descent(x,y):
    m_curr=b_curr=0
    itr=10000000
    n=len(x)
    rate=0.0002
    for i in range(itr):
        y_pred=m_curr*x+b_curr
        cost=(1/n)*sum([val**2 for val in (y-y_pred)])
        print(f'm_prev :{m_curr} , b_prev :{b_curr}, learning rate:{rate} , prev cost :{cost}, Iteration No: {i}')
        md=-(2/n)*sum(x*(y-y_pred))
        bd=-(2/n)*sum(y-y_pred)
        m_curr=m_curr-rate*md
        b_curr=b_curr-rate*bd
        yy_pred=m_curr*x+b_curr
        curr=(1/n)*sum([val**2 for val in (y-yy_pred)])
        print(f'm_curr :{m_curr},b_curr :{b_curr},learning rate : {rate} , Iteration No: {i},curr cost:{curr}')
        if math.isclose(cost,curr,rel_tol=1e-20,abs_tol=0.0):
                return 



    
        
        
```
